# Classitication report metrics

These experiments consider MTA.

```python
import glob
import torch

import numpy as np
import pandas as pd

from utils import get_pred_and_real, get_classification_report
```

```python
path_root = 'results_paper'

# Same order that in dataset.py 
classes = [ 'AGN', 'CART', 'Cepheid', 'Delta Scuti', 'Dwarf Novae', 'EB', 'ILOT',
            'KN', 'M-dwarf Flare', 'PISN', 'RR Lyrae', 'SLSN', '91bg', 'Ia', 'Iax', 'Ib/c',
            'II', 'SN-like/Other', 'TDE', 'uLens']

# Order that we want
labels_order = ['CART', 'Iax', '91bg', 'Ia', 'Ib/c', 'II', 'SN-like/Other', 'SLSN', 
                'PISN', 'TDE', 'ILOT', 'KN', 'M-dwarf Flare', 'uLens', 'Dwarf Novae', 
                'AGN', 'Delta Scuti', 'RR Lyrae', 'Cepheid', 'EB']
```

## ATAT (LC + MD + Features)

```python
path_lc_md_feat_mta = 'lc_tab/Exp_cfg_-arch=lc_header_feat_mtime'
```

```python
test_proba_model, df_y_test = get_pred_and_real(path_root, path_lc_md_feat_mta, classes)

num_folds = len(test_proba_model)
print('Num_folds: {}'.format(num_folds))

display(test_proba_model[0].shape)
display(test_proba_model[1].shape)
display(df_y_test.shape)
```

```python
dict_mean_classes, dict_std_classes, report_metrics = get_classification_report(test_proba_model, df_y_test, num_folds)
dict_mean_classes.keys()
```

```python
max_label_length = max(len(label) for label in labels_order)
max_value_length = max(max(len(f"{dict_mean_classes[metric][label]}") for label in labels_order) for metric in ['f1-score', 'precision', 'recall'])
max_std_length = max(max(len(f"{dict_std_classes[metric][label]}") for label in labels_order) for metric in ['f1-score', 'precision', 'recall'])

for metric in ['precision', 'recall', 'f1-score']:
    print(f'#------------ {metric.upper()} ------------#')

    for label in labels_order:
        mean_val = f"{dict_mean_classes[metric][label]:>{max_value_length}}"
        std_val = f"{dict_std_classes[metric][label]:<{max_std_length}}"
        print(f'{label:<{max_label_length}}: {mean_val} + {std_val}')
```

```python
for metric in ['precision', 'recall', 'f1-score']:

    list_mean_class = []
    list_std_class = []
    for i_fold in range(num_folds):
        list_mean_class.append(report_metrics[i_fold]['macro avg'][metric] * 100)
        list_std_class.append(report_metrics[i_fold]['macro avg'][metric] * 100)

    list_mean_class = round(np.mean(list_mean_class), 2)
    list_std_class = round(np.std(list_std_class), 2)

    print('Macro avg ({}): {} + {}'.format(metric.upper(), list_mean_class, list_std_class))
```

## Random Forest

```python
path_root = 'results_rf_paper/2048_days'
```

```python
test_proba_model = []
num_folds = 5

for i_fold in range(num_folds):
    test_proba_model.append(pd.read_parquet('./{}/test_predictions_proba_model_{}.parquet'.format(path_root, i_fold)))

    for idx_row in range(test_proba_model[i_fold].shape[0]):
        idx_col = test_proba_model[i_fold].iloc[idx_row].argmax()
        test_proba_model[i_fold].iloc[idx_row][idx_col] = 1

    test_proba_model[i_fold][test_proba_model[i_fold] != 1] = 0

display(test_proba_model[0].head())
display(test_proba_model[1].head())
print(test_proba_model[0].shape)
```

```python
test_set = pd.read_parquet('./data_partition_paper/partitions.parquet')
test_set = test_set[test_set.partition == 'test'].drop(['partition'], axis=1)
test_set['SNID'] = test_set['SNID'].apply(lambda row: row+'_2048')
test_set = test_set.set_index('SNID')
test_set = pd.get_dummies(test_set.label, dtype=float)
test_set = test_set[test_proba_model[0].columns]
test_set
```

```python
dict_mean_classes, dict_std_classes, report_metrics = get_classification_report(test_proba_model, test_set, num_folds)
dict_mean_classes.keys()
```
